```python
from extraerNumeros import extraerNumeros

# Agregar el nombre del archivo en la siguiente variable
archivo = '.csv'
ri = extraerNumeros(archivo)

def jugadaPar(numeros):
    return numeros[0] == numeros[1] or numeros[1] == numeros[2] or numeros[2] == numeros[3] or numeros[3] == numeros[4] 

def jugadaDoblePar(numeros):
    return (numeros[0] == numeros[1] and numeros[2] == numeros[3]) or (numeros[0] == numeros[1] and numeros[3] == numeros[4]) or (numeros[1] == numeros[2] and numeros[3] == numeros[4])

def jugadaTercia(numeros):
    return (numeros[0] == numeros[1] and numeros[1] == numeros[2]) or (numeros[1] == numeros[2] and numeros[2] == numeros[3]) or (numeros[2] == numeros[3] and numeros[3] == numeros[4])

def jugadaTerciaPar(numeros):
    return (numeros[0] == numeros[1] and numeros[1] == numeros[2] and numeros[3] == numeros[4]) or (numeros[0] == numeros[1] and numeros[2] == numeros[3] and numeros[3] == numeros[4])

def jugadaPoker(numeros):
    return (numeros[0] == numeros[1] and numeros[1] == numeros[2] and numeros[2] == numeros[3]) or (numeros[1] == numeros[2] and numeros[2] == numeros[3] and numeros[3] == numeros[4])

def jugadaQuintila(numeros):
    return numeros.count(numeros[0]) == 5

def ordenar(numero):
    for x in range(1, len(numero)):
        for y in range(len(numero) - x):
            if numero[y + 1] < numero[y]:
                aux = numero[y]
                numero[y] = numero[y + 1]
                numero[y + 1] = aux

def pruebaPoker(conjuntoNums): 
    numeros = []
    for x in range(len(conjuntoNums)):
        if len(str(conjuntoNums[x])[2:7]) == 1:
            num = str(conjuntoNums[x])[2:7] + '0000'
        elif len(str(conjuntoNums[x])[2:7]) == 2:
            num = str(conjuntoNums[x])[2:7] + '000'
        elif len(str(conjuntoNums[x])[2:7]) == 3:
            num = str(conjuntoNums[x])[2:7] + '00'
        elif len(str(conjuntoNums[x])[2:7]) == 4:
            num = str(conjuntoNums[x])[2:7] + '0'
        else:
            num = str(conjuntoNums[x])[2:7]
        numeros.append(num)

    orden = []
    td = 0
    p = 0
    dp = 0
    tp = 0
    t = 0
    poker = 0
    q = 0
    counter = 0

    f1 = []
    f2 = []
    f3 = []
    f4 = []
    tablas = 12.59

    for x in range(len(numeros)):
        for y in range(5):
            orden.append(int(numeros[x][y]))
        ordenar(orden)

        if jugadaQuintila(orden):
            q += 1
        elif jugadaPoker(orden):
            poker += 1
        elif jugadaTerciaPar(orden):
            tp += 1
        elif jugadaTercia(orden):
            t += 1
        elif jugadaDoblePar(orden):
            dp += 1
        elif jugadaPar(orden):
            p += 1
        else:
            td += 1
        counter +=1
        orden = []

    f1.append(0.3024 * 10000)
    f1.append(0.5040 * 10000)
    f1.append(0.1080 * 10000)
    f1.append(0.0090 * 10000)
    f1.append(0.0720 * 10000)
    f1.append(0.0045 * 10000)
    f1.append(0.0001 * 10000)

    f2.append(f1[0] - td)
    f2.append(f1[1] - p)
    f2.append(f1[2] - dp)
    f2.append(f1[3] - tp)
    f2.append(f1[4] - t)
    f2.append(f1[5] - poker)
    f2.append(f1[6] - q)

    f3.append(f2[0] ** 2)
    f3.append(f2[1] ** 2)
    f3.append(f2[2] ** 2)
    f3.append(f2[3] ** 2)
    f3.append(f2[4] ** 2)
    f3.append(f2[5] ** 2)
    f3.append(f2[6] ** 2)

    f4.append(f3[0] / f1[0])
    f4.append(f3[1] / f1[1])
    f4.append(f3[2] / f1[2])
    f4.append(f3[3] / f1[3])
    f4.append(f3[4] / f1[4])
    f4.append(f3[5] / f1[5])
    f4.append(f3[6] / f1[6])

    return sum(f4) < tablas

if pruebaPoker(ri):
    print('El conjunto de numeros ri pasaron la prueba de poker')
else:
    print('El conjunto de numeros ri no pasaron la prueba de poker')
```
